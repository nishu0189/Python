### Assignment 6: List Methods

Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

````python
import random

rand_list = [random.randint(1,20) for _ in range(15)]  #create rand num list
print(f"Original List: {rand_list}")  #print rand_list

asc_Sort = sorted(rand_list)  # sort asc
print(f"Sort list in asc order:{asc_Sort}")

dsc_Sort = sorted(rand_list, reverse=True) # sort desc 
print(f"Sort list in desc order:{dsc_Sort}")

rm_dup_list = list(set(rand_list))
print(f"Remove duplicate from list: {rm_dup_list}")
````

### Assignment 9: Matrix Transposition

Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

````python
def transpose_matrix(matrix):
    transpose = []

    for i in range(3):
        row = []
        for j in range(3):
            row.append(matrix[j][i])
        transpose.append(row)

    return transpose

matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
    
transposed = transpose_matrix(matrix)
print("Original matrix: ")
for row in matrix:
    print(row)
print("Transposed maatrix: ")
for row in transposed:
    print(row)
````

### Assignment 10: Flattening a Nested List

Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

````python
def to_flatten(nst_list):
    flat_list = []
    for i in nst_list:
        for j in nst_list:
            flat_list.append(j)
    return flat_list
            
        
nested_list = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

x = to_flatten(nested_list)
print("Flatten list: ",x)
    
````

### Assignment 13: List Reversal

Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists

````python
def reverse_list(lst):
    return lst[::-1]

original_lst = [1,2,3,4,5]
rev_list= reverse_list(original_lst)
print(f"Original list: {original_lst}")
print(f"Reverse list: {rev_list}")
````

### Assignment 14: List Rotation

Write a function that rotates a list by n positions. Print the original and rotated lists.

````python
def reverse_n_pos(lst,n):
    return lst[n:] + lst[:n]
    
lst= [1,2,3,4,5]
a = reverse_n_pos(lst,3)
print(f"Original list: {lst}")
print(f"Rotated list: {a}")
````

### Assignment 15: List Intersection

Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

````python
def common_ele(lst1,lst2):
    return [x for x in lst1 if x in lst2]

lst1 = [1,2,3,4,5]
lst2 = [1,5,6,3,8]

intersection = common_ele(lst1,lst2)
print(f"List 1: {lst1}")
print(f"List 2: {lst2}")
print(f"Intersection: {intersection}")
````
