## NOTE Immutability of tuples
Immutability means once a tuple is created, its contents cannot be changed.
In Python, tuples are immutable, unlike lists which are mutable.

**After creating a tuple, you cannot:**
- Change an element
- Add new elements (no append , no insert)
- Remove existing elements

Since tuples are immutable, you cannot change them directly.
But if you still want to “modify” a tuple, you can do it indirectly using these methods:-
- Convert tuple into a list
- Make changes in the list
- Convert the list back into a tuple

### Assignment 15: Tuple Functions

Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

```python
tup = (1,2,3,4,5,100)

def find_min(tup):
    return min(tup)
    
def find_max(tup):
    return max(tup)
    
def find_sum(tup):
    return sum(tup)

print("Minimum element: ",find_min(tup))
print("Maximum element: ",find_max(tup))
print("Sum of  element: ",find_sum(tup))
```
