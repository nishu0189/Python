# Real World Examples using list in python

##  EXAMPLE 1. Manage a To-Do-List

```python
to_do_list = ["Meditation","Complete assignments","Skin Care","Schedule meeting"]

# Add task to list
to_do_list.append("Make Project")
to_do_list.append("Pay Bills")

# Removing completed task
to_do_list.remove("Complete assignments")
to_do_list.remove("Make Project")

# Check if the task is in the list and give reminder
if "Pay Bills" in to_do_list:
    print("Reminder: Pay your utility bills")

#print the remaining task
print("Renaming task: ")
for i in to_do_list:
    print(i)
```

## EXAMPLE 2. Organising students Marks
- create list to store and calculate the average marks for students

```python
# organising students marks
marks = [80,48,92,65,98]

# Adding a new marks
marks.append(88)

# Calculating the avg marks
avg_marks = sum(marks)/len(marks)
print("Average Marks :",avg_marks)

# Finding the highest and lowest marks
highest_marks = max(marks)
lowest_marks = min(marks)

print("Highest Marks: ", highest_marks)
print("Lowest Marks: ", lowest_marks)
```

## EXAMPLE 3. Managing As Inventory
Use a list to manage inventory items in store

```python
inventory = ["pineapple","apples","bananas","Mangoes"]

# add new items
inventory.append("Strawberry")

# Removing the iteam that is out of stock
inventory.remove("Mangoes")

#checking if an item is in stock
item = "oranges"
if item in inventory:
    print(f"{item} is in stock")
else:
    print(f"{item} is out of stock")

# printing the inventory

print("Inventory list: ")
for item in inventory:
    print(f"- {item}")
```

## EXAMPLE 4. Collecting User feedback
- Use a list to collect and analyze user feedback.

```python
# Collecting user feedback

feedback=  ["Greate Service !", "very satisfied","greate experience", "could be better", "Excellent experience"]

#add new feedback
feedback.append("Not happy with the service")

#Counting the positive feedback
positive_feedback_cnt = sum(1 for cmt in feedback if "great" in cmt.lower() or "excellent" in cmt.lower())
print("No.of positive feedback count: ",positive_feedback_cnt)

print("User Feedback: ")
for cmt in feedback:
    print(f"- {cmt}")
```
