# Lambda Function
A lambda function in python is small, anonymous(nameless) function written in one single line.

lambda **arguments : expression**
- No def
- No function name
- Automatically returns the result

### Normal Function

```
add = lambda a,b:a+b
print(add)
```

### Lambda Function

````python
def add(a,b):
    return a+b

print(add(5,10))
````

### Square of a Number

````python
sqr = lambda x:x**2
print(sqr(5))
````

### Lambda with if-else

````python
check_even = lambda n:"Even" if n%2 == 0 else "Odd"
print(check_even(10))
print(check_even(3))
````

### Lambda With Map
- **Map:** is the build-in python function that applied the same function operation to each element of an iterable(like a tuple,list etc) and returns the result.
- **Syntax:** map(**function, iterable**)


````python
num = [1,2,3,4,5]
result = list(map(lambda x:x**2,num))
print(result)
````

````python
n = [1,2,3,4]
list(map(lambda a:a+10,n))
````

Map with Filter

````python
nums = [1, 2, 3, 4, 5]
even_nums = list(filter(lambda x: x % 2 == 0, nums))
print(even_nums)
````
