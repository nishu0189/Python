# ARGUMENT

### 1. Default Parameter in function

````python
def greet(name="viewer"):
          print(f"Hello {name}, Welcome to my channel")

greet("Nishu")
greet() # default parameter run
````

### 2. Variable length Parameter (POSITIONAL)
Variable length parameter allow a function to accept any no. of arguments, instead of fixed count.

**syntax:**

- def func_name(*args)

````python
def print_val(*args):
    for i in args:
        print(i)

print_val(1,52,6,237,"xyz",24,35,'a',"hee",'c',)
````

````python
def addition(*n):
    total=0
    for i in n:
        total += i
        
    return total

print("Addition: ",addition(1,2,3))
print("Addition: ",addition(10,20,20,50,100))
````

### 2.1 Variable-Length Keyword Parameters (**kwargs) (KEYWORD)
**kwargs allows a function to accept any number of keyword (name = value) arguments.

**syntax:**
- def func_name(*kwargs)

````python
def std_details(**kwargs):
    print(kwargs)

std_details(name="Nishu", age=21, course="Python")
````

````python
def std_details(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)

std_details(name="Nishu", age=21, course="Python")
````

## 3. Positional 
Positional arguments are function parameters that recieve values according to the order of argument passed during function call. 

**syntax:**
- def func_name(x,y,z)

````python
def multi(x,y,z):
    return x*y*z

multi(1,2,3)
````

````python
def greet(name,msg):
    print(msg,name)

# positional arguments
greet("Nishu","Hello")

# keyword arguments 
greet(msg="Hello",name="Avi")
````

### 4. Keyword parameter
The keyword parameter is a function parameter that recieves its value by parameter name, not by position

````python
def student(name, age):
    print(name, age)

student(age=22, name="Nishu")
````

## Combination of Positional and Keyword Argument

````python
def print_val(*args,**kwargs):
    for i in args:
        print(f"Positional Argument: {i}")

    print("Keyword Argument: ")
    for key,value in kwargs.items():
        print(f"{key}:{value}")


print_val(1,2,'xyz',Name="Nishu",Branch="Ai&Ds")        
````

## Multiple Return Parameters

````python
def show(a,b):
    add=a+b
    return add,a,b

show(5,10)
````

````python
def cal(a,b):
    return a+b, a*b, a-b

add, mul, sub = cal(10,5)
print("Addition: ",add)
print("Multiply: ",mul)
print("Subtract: ", sub)
````

# Coding example with Function 

### Example 1: Temperature Conversion

````python
def temp_conversion(temp,unit):
# This function convert temp btw celsius to fahrenheit
    if unit == 'C':
        return temp* (9/5) + 32 # celsius to fahrenheit
    elif unit == 'F':
        return (temp-32)*5/9    # fahrenheit to celsius
    else:
        None

print(f"Celsius to Fahrenheit:{ temp_conversion(30,'C')}")
print(f"Fahrenheit to Celsius:{ temp_conversion(105,'F')}")
````

### Example 2: Password Strength Checker

````python
def is_strong_pass(password):
    if len(password) < 8:
        return False
    elif not any(ch.isdigit() for ch in password):
        return False
    elif not any(ch.islower() for ch in password):
        return False
    elif not any(ch.isupper() for ch in password):
        return False
    elif not any(not ch.isalnum() for ch in password):
        return False
    else:
        return True

password = input("Enter a strong Password")
print(f"Password Checker: {is_strong_pass(password)}")
````

### EXAMPLE 3: Calculate the Total Cost of items in the shopping Cart 

````python
def cal_total_cost(cart):
    total=0
    for item in cart:
        total += item['price']*item['quantity']
    return total

cart = [
    {'name':'Skirt' , 'price':0.8 , 'quantity':1},
    {'name':'Books' , 'price':0.2 , 'quantity':10},
    {'name':'football','price':0.5 ,'quantity':5}
]

print(f"Total cost:{cal_total_cost(cart)}")
````

### EXAMPLE 4: Check if a String is palindrome

````python
def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]

s= "A man a plan a canal Panama"

if is_palindrome(s):
    print("This is palindrome")

else:
    print("This is not palindrome")

print(is_palindrome("The Python"))
````

### EXAMPLE 5: Calculate the Factorial of number using recursion

````python
def fac(n):
    if(n == 0):
        return 1
    else:
        return n*fac(n-1)

n = int(input("Enter a number"))
print(f"Factorial of {n}: {fac(n)}")
````

### EXAMPLE 6: A function to read a file and count the frequency of each word

````python

def cnt_word_freq(file_path):
    freq={}   #Dictionary to store word counts
    with open(file_path, 'r') as myfile:
        for line in myfile:
            words = line.lower().split()
            for word_key in words:
                word_key = word_key.strip(".!,?;:\"'()[]{}") #Removes punctuation from word
                freq[word_key]= freq.get(word_key,0) + 1 #Take the current count of the word (or 0 if it doesnâ€™t exist) and add 1

    return freq

filepath = 'sample_file.txt'
word_freq = cnt_word_freq(filepath)
print(word_freq)
````

### EXAMPLE 7: Validate Email Address

````python
def validate_email(email_id):
    if " " in email_id:
        return False
    if not '@' in email_id:
        return False
    username, domain = email_id.split("@")
    if username == "" or domain == "":
        return False
    if "." not in domain: 
        return False

    return True

email = input("Enter a email Id: ")
print("Valid Email!!" if validate_email(email)==True else "Invalid Email")
````
